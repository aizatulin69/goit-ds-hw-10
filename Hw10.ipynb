```
from surprise import Dataset, SVD, SVDpp, NMF
from surprise.model_selection import GridSearchCV, cross_validate
```

```
data = Dataset.load_builtin('ml-100k')

param_grid_svd = {
    'n_factors': [50, 100],
    'n_epochs': [20, 30],
    'lr_all': [0.002, 0.005],
    'reg_all': [0.02, 0.1],}
gs_svd = GridSearchCV(SVD, param_grid_svd, measures=['rmse', 'mae'], cv=3, n_jobs=-1)
gs_svd.fit(data)
print("Best SVD RMSE:", gs_svd.best_score['rmse'])
print("Best SVD params:", gs_svd.best_params['rmse'])

best_svd = gs_svd.best_estimator['rmse']
```

```
svdpp = SVDpp(n_factors=100, n_epochs=30, lr_all=0.005, reg_all=0.1)
cv_svdpp = cross_validate(svdpp, data, measures=['rmse', 'mae'], cv=3, n_jobs=-1)

print("SVD++ RMSE:", cv_svdpp['test_rmse'].mean())
print("SVD++ MAE:", cv_svdpp['test_mae'].mean())
```

```
nmf = NMF(n_factors=50, n_epochs=50, reg_pu=0.06, reg_qi=0.06)
cv_nmf = cross_validate(nmf, data, measures=['rmse', 'mae'], cv=3, n_jobs=-1)

print("NMF RMSE:", cv_nmf['test_rmse'].mean())
print("NMF MAE:", cv_nmf['test_mae'].mean())
```
